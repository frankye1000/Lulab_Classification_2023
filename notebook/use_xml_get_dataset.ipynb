```python
import os
import cv2 as cv
from glob import glob
import xml.etree.ElementTree as ET

XML_PATH     = "../../img/croped_1s/"
```

```python
train      = ['a','b','c','e','f','g','j','m','n','p','r','s','t','u']
validation = ['d','i','k','o']
test       = ['h','l','q','z']
```

```python
glob(os.path.join(XML_PATH, '*.xml'))
```

```python
TRAIN_PATH      = "../dataset_22video_20221031/train/1/"
TEST_PATH       = "../dataset_22video_20221031/test/1/"
VALIDATION_PATH = "../dataset_22video_20221031/validation/1/"

have_papilla_img_name = []
for xml_path in glob(os.path.join(XML_PATH, '*.xml')):
    img_name = xml_path.split("\\")[1].split(".")[0]
    have_papilla_img_name.append(img_name)
    
    if img_name[0] in train:
        img_array = cv.imread(XML_PATH + img_name + '.png')
        # 兩次雙邊濾波
        # blurred_img  = cv.bilateralFilter(img_array, 15, 50,  50)
        # blurred_img2 = cv.bilateralFilter(blurred_img, 15, 50, 50)
        cv.imwrite(TRAIN_PATH + img_name + '.png' , img_array)
    
    elif img_name[0] in test:
        img_array = cv.imread(XML_PATH + img_name + '.png')
        # 兩次雙邊濾波
        # blurred_img  = cv.bilateralFilter(img_array, 15, 50,  50)
        # blurred_img2 = cv.bilateralFilter(blurred_img, 15, 50, 50)
        cv.imwrite(TEST_PATH + img_name + '.png' , img_array)
    
    elif img_name[0] in validation:
        img_array = cv.imread(XML_PATH + img_name + '.png')
        # 兩次雙邊濾波
        # blurred_img  = cv.bilateralFilter(img_array, 15, 50,  50)
        # blurred_img2 = cv.bilateralFilter(blurred_img, 15, 50, 50)
        cv.imwrite(VALIDATION_PATH + img_name + '.png' , img_array)
```

```python
TRAIN_PATH      = "../dataset_22video_20221031/train/0/"
TEST_PATH       = "../dataset_22video_20221031/test/0/"
VALIDATION_PATH = "../dataset_22video_20221031/validation/0/"

for png_path in glob(os.path.join(XML_PATH, '*.png')):
    img_name = png_path.split("\\")[1].split(".")[0]
    if img_name not in have_papilla_img_name:  # no papilla
    
        if img_name[0] in train:
            img_array = cv.imread(XML_PATH + img_name + '.png')
            # 兩次雙邊濾波
            # blurred_img  = cv.bilateralFilter(img_array, 15, 50,  50)
            # blurred_img2 = cv.bilateralFilter(blurred_img, 15, 50, 50)
            cv.imwrite(TRAIN_PATH + img_name + '.png' , img_array)
        
        if img_name[0] in test:
            img_array = cv.imread(XML_PATH + img_name + '.png')
            # 兩次雙邊濾波
            # blurred_img  = cv.bilateralFilter(img_array, 15, 50,  50)
            # blurred_img2 = cv.bilateralFilter(blurred_img, 15, 50, 50)
            cv.imwrite(TEST_PATH + img_name +'.png' , img_array)
        
        if img_name[0] in validation:
            img_array = cv.imread(XML_PATH + img_name + '.png')
            # 兩次雙邊濾波
            # blurred_img  = cv.bilateralFilter(img_array, 15, 50,  50)
            # blurred_img2 = cv.bilateralFilter(blurred_img, 15, 50, 50)
            cv.imwrite(VALIDATION_PATH + img_name + '.png' , img_array)
```
